```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import classification_report
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
# Raw GitHub URL for the CSV file
url = "https://raw.githubusercontent.com/chiru30/COMP-6321-Machine-Learning/main/major%20assignment%202/Question%204/survey_lung_cancer.csv"

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(url)
df.head()
```

```python
label_encoder = LabelEncoder()
df['GENDER'] = label_encoder.fit_transform(df['GENDER'])
df.head()
```

```python
X = df.drop('LUNG_CANCER', axis=1)
y = df['LUNG_CANCER']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
classifier_reports=[]
```

```python
# (b) Train a K-nearest neighbors (KNN) model for the classification task with K=10
knn = KNeighborsClassifier(n_neighbors=10)
knn.fit(X_train, y_train)
knn_train_pred = knn.predict(X_train)
knn_test_pred = knn.predict(X_test)
print("KNN Classification Report on Training Set:")
print(classification_report(y_train, knn_train_pred))
print("KNN Classification Report on Testing Set:")
print(classification_report(y_test, knn_test_pred))
knn_report = classification_report(y_test, knn_test_pred, output_dict=True)
classifier_reports.append({'KNN': knn_report})
```

```python
import warnings
warnings.filterwarnings('ignore')
# (c) Train a Support Vector Machine (SVM) model for the classification task
svm = SVC()
svm.fit(X_train, y_train)
svm_train_pred = svm.predict(X_train)
svm_test_pred = svm.predict(X_test)
print("SVM Classification Report on Training Set:")
print(classification_report(y_train, svm_train_pred))
print("--------------------------------------------------------")
print("SVM Classification Report on Testing Set:")
print(classification_report(y_test, svm_test_pred))
svm_report = classification_report(y_test, svm_test_pred, output_dict=True)
classifier_reports.append({'SVM': svm_report})
```

```python
# (d) Train a Gaussian Naive Bayes (GaussianNB) model for the classification task
gnb = GaussianNB()
gnb.fit(X_train, y_train)
gnb_train_pred = gnb.predict(X_train)
gnb_test_pred = gnb.predict(X_test)
print("Gaussian Naive Bayes Classification Report on Training Set:")
print(classification_report(y_train, gnb_train_pred))
print("--------------------------------------------------------")
print("Gaussian Naive Bayes Classification Report on Testing Set:")
print(classification_report(y_test, gnb_test_pred))
gnb_report = classification_report(y_test, gnb_test_pred, output_dict=True)
classifier_reports.append({'GaussianNB': gnb_report})
```

```python
# (e) Train a Decision Tree model for the classification task
decision_tree = DecisionTreeClassifier()
decision_tree.fit(X_train, y_train)
dt_train_pred = decision_tree.predict(X_train)
dt_test_pred = decision_tree.predict(X_test)
print("Decision Tree Classification Report on Training Set:")
print(classification_report(y_train, dt_train_pred))
print("--------------------------------------------------------")
print("Decision Tree Classification Report on Testing Set:")
print(classification_report(y_test, dt_test_pred))
dt_report = classification_report(y_test, dt_test_pred, output_dict=True)
classifier_reports.append({'DecisionTree': dt_report})
```

```python
# (f) Train an AdaBoost model for the classification task
adaboost = AdaBoostClassifier()
adaboost.fit(X_train, y_train)
ada_train_pred = adaboost.predict(X_train)
ada_test_pred = adaboost.predict(X_test)
print("AdaBoost Classification Report on Training Set:")
print(classification_report(y_train, ada_train_pred))
print("--------------------------------------------------------")
print("AdaBoost Classification Report on Testing Set:")
print(classification_report(y_test, ada_test_pred))
ada_report = classification_report(y_test, ada_test_pred, output_dict=True)
classifier_reports.append({'AdaBoost': ada_report})
```

```python


class ClassifierReports:
    def __init__(self, url):
        self.df = pd.read_csv(url)
        self.label_encoder = LabelEncoder()
        self.df['GENDER'] = self.label_encoder.fit_transform(self.df['GENDER'])
        self.X = self.df.drop('LUNG_CANCER', axis=1)
        self.y = self.df['LUNG_CANCER']
        self.X_train, self.X_test, self.y_train, self.y_test = train_test_split(self.X, self.y, test_size=0.3, random_state=42)
        self.classifier_reports = []

    def train_classifier(self, classifier, name):
        model = classifier()
        model.fit(self.X_train, self.y_train)
        test_pred = model.predict(self.X_test)
        report = classification_report(self.y_test, test_pred, output_dict=True)
        self.classifier_reports.append({name: report})

    def generate_reports(self):
        classifiers = {
            'KNN': KNeighborsClassifier,
            'SVM': SVC,
            'GaussianNB': GaussianNB,
            'DecisionTree': DecisionTreeClassifier,
            'AdaBoost': AdaBoostClassifier
        }

        for clf_name, clf_class in classifiers.items():
            self.train_classifier(clf_class, clf_name)

    def get_comparative_report(self):
        comparative_classification_report = pd.DataFrame(columns=['Classifier', 'Accuracy', 'Precision', 'F1-Score'])

        for report_dict in self.classifier_reports:
            for clf_name, report in report_dict.items():
                accuracy = report['accuracy']
                precision = report['macro avg']['precision']
                f1_score = report['macro avg']['f1-score']
                comparative_classification_report = comparative_classification_report.append({
                    'Classifier': clf_name,
                    'Accuracy': accuracy,
                    'Precision': precision,
                    'F1-Score': f1_score
                }, ignore_index=True)

        return comparative_classification_report

# Create an instance of ClassifierReports
url = "https://raw.githubusercontent.com/chiru30/COMP-6321-Machine-Learning/main/major%20assignment%202/Question%204/survey_lung_cancer.csv"
classifier_reports_handler = ClassifierReports(url)

# Generate classification reports for various classifiers
classifier_reports_handler.generate_reports()

# Get the comparative classification report DataFrame
comparative_report_df = classifier_reports_handler.get_comparative_report()
print(comparative_report_df)
```

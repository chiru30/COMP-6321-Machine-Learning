```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

label_encoder = LabelEncoder()

df['GENDER'] = label_encoder.fit_transform(df['GENDER'])

df.head()
```

```python
X = df.drop('LUNG_CANCER', axis=1)
y = df['LUNG_CANCER']
```

```python
scaler = StandardScaler()
X['AGE'] = scaler.fit_transform(X[['AGE']])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
import warnings
warnings.filterwarnings('ignore')

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score

logistic_regression = LogisticRegression(random_state=0)

logistic_regression.fit(X_train, y_train)
```

```python
# Train set performance
y_train_pred = logistic_regression.predict(X_train)
train_report = classification_report(y_train, y_train_pred, target_names=['NO', 'YES'])
print("Train Set Classification Report:\n", train_report)

train_accuracy = accuracy_score(y_train, y_train_pred)
train_error = 1 - train_accuracy

print(f"Training Accuracy: {train_accuracy * 100:.2f}%")
print(f"Training Error: {train_error * 100:.2f}%")
```

```python
# Test set performance
y_test_pred = logistic_regression.predict(X_test)
test_report = classification_report(y_test, y_test_pred, target_names=['NO', 'YES'])
print("Test Set Classification Report:\n", test_report)

test_accuracy = accuracy_score(y_test, y_test_pred)
test_error = 1 - test_accuracy

print(f"Test Accuracy: {test_accuracy * 100:.2f}%")
print(f"Test Error: {test_error * 100:.2f}%")
```
